- **Data:**
Dataset ini berisi informasi mengenai program studi dengan kode, nama, jumlah peminat dari tahun 2021 hingga 2023, serta akreditasinya.

- **Tujuan Visualisasi:**
Untuk menunjukkan tren peminat pada berbagai program studi dari 2021 hingga 2023.
Membandingkan jumlah peminat di berbagai jurusan.
User: Visualisasi ini cocok untuk pemangku kepentingan di bidang pendidikan, seperti admin universitas, calon mahasiswa, dan pihak terkait perencanaan akademik.

- **Fungsi:**
Mempermudah identifikasi jurusan yang mengalami peningkatan atau penurunan jumlah peminat.
Menyediakan informasi visual yang bisa digunakan untuk pengambilan keputusan.
Tone: Informal tapi jelas, sesuai untuk presentasi dan laporan internal.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("datasetssd.csv", sep=";")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop('AKREDITASI',axis=1)
```

```python
dfTransaction = df.groupby('NAMA')['PEMINAT 2021'].sum().reset_index()

# Mengurutkan DataFrame berdasarkan total uang dalam urutan menurun dan mengambil 10 teratas
dfTransaction = dfTransaction.sort_values('PEMINAT 2021', ascending=False).head(10)

# Membuat plot batang
plt.figure(figsize=(12, 8))
plt.bar(data=dfTransaction, x='NAMA', height='PEMINAT 2021', color='pink')
plt.title('10 Prodi dengan peminat terbanyak di ITERA pada tahun 2021')
plt.xticks(rotation=45)
plt.xlabel('Nama Prodi')
plt.ylabel('Banyak Peminat')
plt.show()
```

```python
dfTransaction = df.groupby('NAMA')['PEMINAT 2022'].sum().reset_index()

# Mengurutkan DataFrame berdasarkan total uang dalam urutan menurun dan mengambil 10 teratas
dfTransaction = dfTransaction.sort_values('PEMINAT 2022', ascending=False).head(10)

# Membuat plot batang
plt.figure(figsize=(12, 8))
plt.bar(data=dfTransaction, x='NAMA', height='PEMINAT 2022', color='lightblue')
plt.title('10 Prodi dengan peminat terbanyak di ITERA pada tahun 2022')
plt.xticks(rotation=45)
plt.xlabel('Nama Prodi')
plt.ylabel('Banyak Peminat')
plt.show()
```

```python
df.sort_values(by='PEMINAT 2023', ascending=False)
```

```python
dfTransaction = df.groupby('NAMA')['PEMINAT 2023'].sum().reset_index()

# Mengurutkan DataFrame berdasarkan total uang dalam urutan menurun dan mengambil 10 teratas
dfTransaction = dfTransaction.sort_values('PEMINAT 2023', ascending=False).head(10)

# Membuat plot batang
plt.figure(figsize=(12, 8))
plt.bar(data=dfTransaction, x='NAMA', height='PEMINAT 2023', color='lightblue')
plt.title('10 Prodi dengan peminat terbanyak di ITERA pada tahun 2023')
plt.xticks(rotation=45)
plt.xlabel('Nama Prodi')
plt.ylabel('Banyak Peminat')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df['KODE'], df['PEMINAT 2022'], marker='o', color='red', linestyle='-', linewidth=2, markersize=8)

# Menambahkan judul dan label
plt.title('DATA PEMINAT PRODI DI ITERA 2022')
plt.xlabel('KODE PRODI (1920..)')
plt.ylabel('PEMINAT 2022')
plt.xticks(rotation=45)

# Menampilkan grafik
plt.show()
print(df[['NAMA', 'KODE']] )
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df['KODE'], df['PEMINAT 2023'], marker='o', color='red', linestyle='-', linewidth=2, markersize=8)

# Menambahkan judul dan label
plt.title('DATA PEMINAT PRODI DI ITERA 2023')
plt.xlabel('KODE PRODI (1920..)')
plt.ylabel('PEMINAT 2023')
plt.xticks(rotation=45)

# Menampilkan grafik
plt.show()
print(df[['NAMA', 'KODE']] )
```

```python

# Data preparation for plotting
data_melted = df.melt(id_vars=["NAMA"],
                        value_vars=["PEMINAT 2021", "PEMINAT 2022", "PEMINAT 2023"],
                        var_name="Tahun",
                        value_name="Peminat")

# Plotting
plt.figure(figsize=(12, 8))
sns.lineplot(data=data_melted, x="Tahun", y="Peminat", hue="NAMA", marker="o")

# Customize the plot
plt.title("Tren Jumlah Peminat Program Studi 2021-2023", fontsize=16)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Jumlah Peminat", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Program Studi", bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
